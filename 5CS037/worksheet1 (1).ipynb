```python
#Solve the following system of Linear equation using Inverse Methods.2x − 3y + z = −1,x − y + 2z = −3,3x + y − z = 9

A2 = np.array([[ 2, -3,  1],[ 1, -1,  2],[ 3,  1, -1]])

# Constant matrix B
B2 = np.array([-1, -3, 9])

print("Coefficient Matrix A2 =\n", A2)
print("Constant Matrix B2 =\n", B2)
print()

A2_inv = np.linalg.inv(A2)

X = np.dot(A2_inv, B2)

print("Solution using Inverse Method (X = A^-1 · B):")
print("x =", X[0])
print("y =", X[1])
print("z =", X[2])
```

```python
#Matrix Multiplication
#Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.
# Python List Matrix Multiplication

import random
import time

# Generating 1000x1000 matrix using lists
A = [[random.random() for _ in range(1000)] for _ in range(1000)]
B = [[random.random() for _ in range(1000)] for _ in range(1000)]

# Function to multiply two matrices using lists
def python_matrix_multiply(A, B):
    result = [[0]*len(B[0]) for _ in range(len(A))]
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    return result

start = time.time()
result_list_matrix = python_matrix_multiply(A, B)
end = time.time()

print("Python List Matrix Multiplication Time:", end - start, "seconds")
```

```python
#Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000. Mea-sure and print the time taken for this operation.

# NumPy Matrix Multiplication

A_np = np.random.rand(1000, 1000)
B_np = np.random.rand(1000, 1000)

start = time.time()
result_np_matrix = np.dot(A_np, B_np)
end = time.time()

print("NumPy Matrix Multiplication Time:", end - start, "seconds")
```
